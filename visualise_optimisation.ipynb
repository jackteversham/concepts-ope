```python
import matplotlib.pyplot as plt

# Read data from the text file
all_learning_rate_data = []
for i in [1,3,5]:
    filename = f"optimisation_histories/a00{i}_d200_n100.txt"
    data = []
    with open(filename, "r") as file:
        for line in file:
            if "|" in line:
                epoch, mse = line.split("|")
                epoch = int(epoch.strip())
                mse = float(mse.strip())
                data.append((epoch, mse))
    all_learning_rate_data.append(data)

# Separate data into separate lists for plotting

fig = plt.figure(figsize=(7,5), dpi=200)

labels = [0.001, 0.003, 0.005]
for i,data in enumerate(all_learning_rate_data):
    epochs, mse_values = zip(*data)
    plt.plot(epochs, mse_values, label=r"$\alpha$"+f"={labels[i]}")

plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.title("MSE over Epochs")
plt.grid(True)
plt.show()
```
